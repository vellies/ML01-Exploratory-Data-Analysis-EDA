```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
#To Check Data Types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
## Pie Chart - Top 3 countries that uses zomota
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

Obsevation: Zomato maximum records or transaction are from india After that USA and then United Kingdoms

```python
### Numerical variables
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observations
1. When rating is between 4.5 to 4.9 ---> Excellent
2. When rating is between 4.0 to 4.4 ---> Very Good
3. When rating is between 3.5 to 3.9 ---> Good
4. When rating is between 3.0 to 3.4 ---> Average
5. When rating is between 2.5 to 2.9 ---> Average
6. When rating is between 1.8 to 2.4 ---> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(data=ratings,x="Aggregate rating",y="Rating Count")
```

```python
sns.barplot(data=ratings,x="Aggregate rating",y="Rating Count",hue="Rating color",palette=['white','red','orange','yellow','green','green'])
```

## Observation
    1. Not Rated count is very high
    2. Maximum number of ratings are between 2.4 to 3.4

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
### Find the countries name that has given 0 ratings
final_df.columns
```

```python
white_countries=final_df[final_df['Rating color']=='White']
white_countries
```

```python
zero_rate_countries=white_countries.groupby(['Country']).size().reset_index().rename(columns={0:"count"})
zero_rate_countries.head()
```

```python
sns.barplot(data=zero_rate_countries,x="Country",y="count")
```

Observations Maximum number of 0 ratings are from indian customers
